```python
#Q-Q plots
from scipy.special import ndtri
#df = df.sort_values(by=['SepalLengthCm'], ascending = True).reset_index()
#df['count']=df.index + 1
df.head()
n_rows = df.shape[0]
df['%area']=(df['count'])/n_rows
df['z_theoretical'] = ndtri(df['%area'])
df['z_actual']=(df['SepalLengthCm'] - df['SepalLengthCm'].mean())/df['SepalLengthCm'].std(ddof=0)
df.head()
```

```python
plt.scatter(df.z_actual,df.z_theoretical)
plt.plot([-2,-1,0,1,2],[-2,-1,0,1,2])
```

Q-Q plot, short for quantile-quantile plot, is basically a tool used for understanding the data sets we have. We can know the closet approximation that we can assume fot our data sets by comparing to distributions. Majorly its used to 
know about the relation between the data sets and normal distribution, but it can be compared to other distributions too.
  Hence we can understand from the above plot that the scatter plot of the data set does not exactly follow the normal distribution but to some extent.
